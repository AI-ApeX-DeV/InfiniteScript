- Types of operating systems
  - batch , multiprogrammed, multitasking, realtime, distributed, clustered and embedded (washing machine)
- Process state diagram 
  - new, ready, running, wait, terminate, suspend wait  / long term scheduler (from new to ready), short term scheduler (from ready to running)
- Some linux commands 
  - Authorization grant 
    - read -> 4 write -> 2 execute -> 1 
    - chmod 666 (means rw right given to user,group and others)
    - chmod u+r
    - chmod u+r,g+w,o+e
    - chmod ugo=r
    - chmod ugo +r
  - movement of disk head
    - lseek(n,5,seek_cur) -> moves the pointer 5 digits ahead eg: 123456  lseek will move pointer from 1 to 5
    - lseek(n,5,seek_set) -> moves the pointer to digit 5 eg: 123456  lseek will move pointer to 5
- System Calls
  - File related (open,read,write,close,create file)
  - Device related (read,write,reposition,ioctl,fcntl)
  - Information (get pid,attributes,get system time and data)
  - Process control (load,execute,abort,fork,wait,signal,allocate)
  - Communication (pipe(),craete/delete connections,shrnget())

-   System call
  - User mode (mode bit = 1 ) and kernel mode (mode bit = 0 )
